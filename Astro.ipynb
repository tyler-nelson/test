```python
import matplotlib.pyplot as plt
import math as math
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from __future__ import division
import pandas as pd
sunColor = plt.cm.inferno(200)
earthColor = plt.cm.inferno(126)
dot = plt.cm.inferno(0)

def constructor(numberOfTestMasses):
    massItems = pd.DataFrame(columns = ['r', 'theta', 'phi', 'rDot', 'thetaDot', 'phiDot', 'ecc', 'mass', 'radius', 'color'])
    rows_list = []
    Sun = {'r':  0.0, 'theta': math.pi/2,'phi': 0, 'rDot': 0.0, 'thetaDot': 0,'phiDot': 0, 'ecc': 0,'mass': 1e10,'radius': 100, 'color': sunColor }
    Earth = {'r':  600, 'theta': math.pi/2,'phi': 0, 'rDot': 0, 'thetaDot': 0,'phiDot': 5, 'ecc': 0,'mass': 1e4,'radius': 20, 'color':earthColor }
    rows_list.append(Sun)
    rows_list.append(Earth)
    for i in range(0, int(numberOfTestMasses)):
        phi = i * 2*math.pi/numberOfTestMasses
        
        temp = {'r':  600 + 5*math.cos(phi), 'theta': math.pi/2,'phi': phi, 'rDot': 0, 'thetaDot': 0,'phiDot': 5, 'ecc': 0,'mass': 1,'radius': 1, 'color': dot}
        rows_list.append(temp)
    '''
    for i in range(0, int(numberOfTestMasses/2)):
        phi = (i + 1) * (2*math.pi - 40/600)/(numberOfTestMasses/2) 
        temp = {'r':  600, 'theta': math.pi/2,'phi': phi, 'rDot': 0, 'thetaDot': 0,'phiDot': 5, 'ecc': 0,'mass': 1,'radius': 1, 'color': dot }
        rows_list.append(temp)
    '''#figure the earth stuff out later
    massItems = pd.DataFrame(rows_list)
    return massItems

G = 6.67*10**(-11)
class SolarSystem2:
    def __init__(self, shape = [1000, 1000, 10], littleRPassed = 0, numberOfTestMasses = 20, twoD = False):
        self.xLim = shape[0]
        self.yLim = shape[1]
        self.zLim = shape[2]
        """#self.fig = plt.figure()
        if(twoD):
            self.ax = self.fig.add_subplot(111)
            self.ax.set_xlim((0, self.xLim))
            self.ax.set_ylim((0, self.yLim))
        else:
            self.ax = self.fig.add_subplot(111, projection = '3d')
            self.ax.set_xlim((0, self.xLim))
            self.ax.set_ylim((0, self.yLim))
            self.ax.set_zlim((0, self.zLim))
        """
        self.x = self.xLim/2
        self.y = self.yLim/2
        self.z = self.zLim/2
        #self.massList = []
        
        data = constructor(numberOfTestMasses)
        data.loc[:,'x'] = pd.Series(data.r*np.sin(data.theta)*np.cos(data.phi), index = data.index) + self.x
        data.loc[:,'y'] = pd.Series(data.r*np.sin(data.theta)*np.sin(data.phi), index = data.index) + self.y
        data.loc[:,'z'] = pd.Series(data.r*np.cos(data.theta), index = data.index) + self.z
        data.loc[:, 'L'] = pd.Series(data.mass*data.r**2*data.phiDot, index = data.index)
        self.totalMass = data['mass'].sum()
        self.system = data
        self.centerOfMass = (1/self.totalMass)*(data.mass*data.r).sum()
    
    def update(self, dt = 0.1):
        for n in self.system[self.system['mass'] < 1e10].itertuples():
            r = n[6]
            theta = n[9]
            phi = n[4]
            rDot = n[4]
            thetaDot = n[10]
            phiDot = n[5]
            mass = n[3]
            acceleration = (-G * self.totalMass)/(self.centerOfMass - r)**2 #+ (mass*r*phiDot)**2/((mass*self.totalMass)/(mass + self.totalMass)*r**3)/mass
            if(r + acceleration*dt < 0):
                self.system.loc[n[0], 'r'] = abs(r + acceleration*dt)
                self.system.loc[n[0], 'theta'] += thetaDot*dt
                self.system.loc[n[0], 'phi'] += phiDot*dt + math.pi/2
            else: 
                self.system.loc[n[0], 'r'] += acceleration*dt
                self.system.loc[n[0], 'theta'] += thetaDot*dt
                self.system.loc[n[0], 'phi'] += phiDot*dt
            self.system.loc[n[0], 'rDot'] += acceleration*dt
            self.system.loc[n[0], 'phiDot'] = n[14]/(n[3]*self.system.loc[n[0], 'r']**2)
            #put in conservation of L for updating phiDot and thetaDot
            
        self.system.loc[:,'x'] = pd.Series(self.system.r*np.sin(self.system.theta)*np.cos(self.system.phi), index = self.system.index) + self.x
        self.system.loc[:,'y'] = pd.Series(self.system.r*np.sin(self.system.theta)*np.sin(self.system.phi), index = self.system.index) + self.y
        self.system.loc[:,'z'] = pd.Series(self.system.r*np.cos(self.system.theta), index = self.system.index) + self.y
        
    def combine(self, radius):
        for n in self.system.itertuples():
            x = n[11]
            y = n[12]
            z = n[13]
            local = self.system.copy()
            local['x'] = abs(local['x'] - x)
            local['y'] = abs(local['y'] - y)
            local['z'] = abs(local['z'] - z)
            local.sort_values(['x','y','z'], ascending=[True, True, True])
            dis = math.sqrt((local.iloc[[1]]['x'] - x)**2 + (local.iloc[[1]]['y'] - y)**2 + (local.iloc[[1]]['z'] - z)**2)
            if(dis < radius):
                workaround = local.iloc[[1]].as_matrix()[0]
                #print workaround[2]
                self.system.loc[n[0], 'mass'] += workaround[2]
                self.system.loc[n[0], 'radius'] += workaround[7]
                self.system.loc[n[0], 'L'] += workaround[13]
                #should add average position considerations
                #we have some weird combine bugs
                self.system.drop(self.system.index[local.iloc[[1]].index], inplace = True)
                
    def plot(self, fig):
        ax = fig.add_subplot(111, projection = '3d')
        for n in self.system.itertuples():
            ax.scatter(n[11],n[12],n[13], color = n[1], s = n[8])
            
        
        
            
    
```
